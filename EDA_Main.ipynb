```python
# Type Casting
student['Avg_Daily_Usage_Hours'] = student['Avg_Daily_Usage_Hours'].astype('int')
student['Sleep_Hours_Per_Night'] = student['Sleep_Hours_Per_Night'].astype('int')
student.dtypes
```

```python
# Histogram
y = student['Academic_Level']
plt.hist(y,color = "purple")
plt.title("Histogram chart")
plt.ylabel("Academic_Level")
plt.grid('True', color = "green")
plt.legend(['Academic_Level'])
```

**Insight from above visualization:**

- Undergraduates dominate the dataset – this group is the most represented.

- Graduate-level individuals also form a significant portion, only slightly fewer than undergraduates.

- High school level individuals are very few – showing minimal representation.

--> This suggests the dataset may be skewed towards people pursuing or having completed higher education (undergraduate and graduate).

```python
# Line chart
plt.figure(figsize = (11,7))
sns.lineplot(data=student, x='Most_Used_Platform', y='Age', estimator='mean', ci=None, sort=False, marker = "o", linestyle="dotted", linewidth = "3")
plt.title('Most_Used_Platform as per Age')
plt.xlabel('Most Used Platform')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

**Insight from above visualization:**

- LinkedIn has the oldest average users (~22.9 years):
This aligns with LinkedIn being a professional network used more by working professionals or students nearing graduation.

- LINE has the youngest average users (~19 years):
Possibly indicating LINE's popularity among younger students, especially in East Asia.

- Snapchat and LINE have relatively younger users (<20):
Suggesting these platforms are preferred by teens or early college students.

- Mid-range age (around 21–22) is observed for:

    Twitter, Facebook, VKontakte, WhatsApp, WeChat

- YouTube, TikTok, and Instagram are mostly used by people aged 20–21, showing popularity among younger audiences but not the youngest.

```python
# Bar chart
plt.bar(student['Sleep_Hours_Per_Night'] ,student['Mental_Health_Score'],color="pink")
plt.title("Result of sleep on mental health")
plt.xlabel("Sleep")
plt.ylabel("Mental Health")
plt.grid('True', color = "green")
plt.legend(['Sleep'])
plt.show()
```

**Insight from above visualization:**

- Optimal sleep duration appears to be around 7–8 hours, yielding the best mental health scores.

- Too little sleep (under 6 hours) and too much sleep (9+ hours) may have less positive effects on mental health.

- The relationship is roughly a bell curve (mental health improves with more sleep until 8 hours, then slightly drops).

```python
# Boxplot
y = student['Addicted_Score']
plt.boxplot(y,vert = False)
plt.title("Addicted_Score")
plt.ylabel("Addicted_Score  ")
plt.grid('True', color = "green")
plt.legend(['Addicted_Score'])
```

**Insight from above visualization:**

- Most users are moderately to highly addicted (scores clustered between 5 and 8).

-  Median addiction level is relatively high (7), suggesting that addiction (possibly to technology or social media) is a significant concern in this group.

- Consistent data without large deviations or outliers.
